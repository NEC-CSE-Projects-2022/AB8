```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import os
import h5py
import numpy as np
import cv2
from tqdm import tqdm
from glob import glob
import random
import shutil
```

```python
input_dir = "dataset/data"
output_image_dir = "yolo/images"
output_mask_dir = "yolo/masks"
output_label_dir = "yolo/labels"

os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_mask_dir, exist_ok=True)
os.makedirs(output_label_dir, exist_ok=True)
```

##  Convert .mat to images and masks

```python
for mat_file in tqdm(os.listdir(input_dir)):
    if mat_file.endswith(".mat"):
        mat_path = os.path.join(input_dir, mat_file)
        with h5py.File(mat_path, "r") as f:
            image = np.array(f["cjdata/image"]).T
            mask = np.array(f["cjdata/tumorMask"]).T

            # Normalize image to uint8
            if image.max() > 255:
                image = (image / image.max()) * 255
            image = image.astype(np.uint8)
            mask = (mask > 0).astype(np.uint8) * 255

        img_name = mat_file.replace(".mat", ".png")
        cv2.imwrite(os.path.join(output_image_dir, img_name), image)
        cv2.imwrite(os.path.join(output_mask_dir, img_name), mask)
```

## Generate YOLO labels

```python
def create_yolo_annotation(mask_path, save_path):
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    h, w = mask.shape
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    annotations = []

    for contour in contours:
        x, y, bw, bh = cv2.boundingRect(contour)
        x_center = (x + bw / 2) / w
        y_center = (y + bh / 2) / h
        bw_norm = bw / w
        bh_norm = bh / h
        annotations.append(f"0 {x_center} {y_center} {bw_norm} {bh_norm}")

    with open(save_path, "w") as f:
        f.write("\n".join(annotations))

print("🔹 Creating YOLO labels...")
mask_files = sorted(glob(f"{output_mask_dir}/*.png"))
for mask in tqdm(mask_files):
    base = os.path.basename(mask).replace(".png", ".txt")
    create_yolo_annotation(mask, os.path.join(output_label_dir, base))
```

```python
# Check if images were created
images = sorted(glob(os.path.join(output_image_dir, "*.png")))
print("✅ Total images created:", len(images))
```

# Split train/val/test

```python
random.seed(42)
n_total = len(images)
n_train = int(n_total * 0.6997)
n_val = int(n_total * 0.2699)
n_test = n_total - n_train - n_val

combined = images.copy()
random.shuffle(combined)

train_imgs = combined[:n_train]
val_imgs = combined[n_train:n_train+n_val]
test_imgs = combined[n_train+n_val:]

print(f"🔹 Split counts -> Train: {len(train_imgs)} Val: {len(val_imgs)} Test: {len(test_imgs)}")
```

# Make directories

```python
splits = ["train", "val", "test"]
for s in splits:
    os.makedirs(f"yolo/{s}/images", exist_ok=True)
    os.makedirs(f"yolo/{s}/labels", exist_ok=True)

def copy_split(images_list, split):
    for img_path in images_list:
        base = os.path.basename(img_path)
        label_path = os.path.join(output_label_dir, base.replace(".png", ".txt"))
        shutil.copy(img_path, f"yolo/{split}/images/{base}")
        shutil.copy(label_path, f"yolo/{split}/labels/{base.replace('.png', '.txt')}")

copy_split(train_imgs, "train")
copy_split(val_imgs, "val")
copy_split(test_imgs, "test")

# Verification
train_count = len(os.listdir("yolo/train/images"))
val_count = len(os.listdir("yolo/val/images"))
test_count = len(os.listdir("yolo/test/images"))
print(f"✅ Data split complete: {train_count} train, {val_count} val, {test_count} test")
```
