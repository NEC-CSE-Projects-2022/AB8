```python
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
model = YOLO('best.pt')
```

```python
# Specify the path to your test image
test_image_path = "yolo/test/images/2023.png"
```

```python
# Perform inference
results = model(test_image_path)

# The results are returned as a list, so we need to access the first item
# and then display the image with bounding boxes and labels
result = results[0]  

# Show the result (image with bounding boxes and labels)
result.show()
```
